# Code for Bisection Method\:
The Bisection Method is a bracketing method used to find roots of a continuous function.
It repeatedly divides the interval into two halves and selects the subinterval where the sign of the function changes.
    c = (a+b)/2
  condition : f(a)⋅f(b)<0

```
import math

def f(x):
    return eval(func)

func = input("Enter function f(x): ")

a = float(input("Enter a: "))
b = float(input("Enter b: "))
tol = float(input("Enter tolerance: "))

if f(a) * f(b) >= 0:
    print("Invalid interval!")
else:
    c = None # Initialize c to None
    while (b - a) >= tol:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c

    if c is not None: # Only print root if c was defined by the loop
        print("Root:", c)
    else:
        print("Root could not be found within the given tolerance or interval.")
```

# Code For Regula Falsi Method
This is also a bracketing method, but instead of midpoint, it uses a linear interpolation between two points to estimate the root
c=(af(b)−bf(a))/f(b)-f(a)
.Uses straight line between (a, f(a)) and (b, f(b))
.Faster than bisection in many cases
	​


```
import math

def f(x):
    return eval(func)

func = input("Enter function f(x): ")

a = float(input("Enter a: "))
b = float(input("Enter b: "))
tol = float(input("Enter tolerance: "))

if f(a) * f(b) >= 0:
    print("Invalid interval!")
else:
    c = a
    while True:
        c = (a*f(b) - b*f(a)) / (f(b) - f(a))

        if abs(f(c)) < tol:
            break

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    print("Root:", c)
```

# Code For Gauss Elimination Method
A direct method used to solve linear systems.
It transforms the system into upper triangular form and then uses back substitution.
Ax=B

```
import numpy as np

n = int(input("Enter number of variables: "))

A = []

print("Enter coefficients row-wise (space-separated values per row):")
for i in range(n):
    row = list(map(float, input().split()))
    A.append(row)

print("Enter constants (space-separated values):")
B = list(map(float, input().split()))

A = np.array(A, dtype=float)
B = np.array(B, dtype=float)

for i in range(n):
    for j in range(i+1, n):

        if A[i][i] == 0:
            print("Error: Division by zero encountered. Gaussian elimination cannot proceed.")
            exit()
        ratio = A[j][i] / A[i][i]
        A[j] -= ratio * A[i]
        B[j] -= ratio * B[i]

x = np.zeros(n)
for i in range(n-1, -1, -1):
    # Check for division by zero
    if A[i][i] == 0:
        print("Error: Division by zero during back substitution. No unique solution or singular matrix.")
        exit()
    x[i] = (B[i] - sum(A[i][j]*x[j] for j in range(i+1, n))) / A[i][i]

print("Solution:", x)
```

# Code For Jacobi's Method
An iterative method where each variable is computed using values from the previous iteration.

```
import numpy as np

n = int(input("Enter number of variables: "))

A = np.zeros((n,n))
B = np.zeros(n)

print("Enter coefficient matrix:")
for i in range(n):
    A[i] = list(map(float, input().split()))

print("Enter constants:")
for i in range(n):
    B[i] = float(input())

x = np.zeros(n)
tol = float(input("Enter tolerance: "))

while True:
    x_new = np.zeros(n)

    for i in range(n):
        s = sum(A[i][j]*x[j] for j in range(n) if j != i)
        x_new[i] = (B[i] - s)/A[i][i]

    if np.allclose(x, x_new, atol=tol):
        break

    x = x_new

print("Solution:", x)
```

# Code For Gauss Siedal Method
Improvement over Jacobi method.
Uses updated values immediately during iteration.

```
import numpy as np

n = int(input("Enter number of variables: "))

A = np.zeros((n,n))
B = np.zeros(n)

print("Enter coefficient matrix:")
for i in range(n):
    A[i] = list(map(float, input().split()))

print("Enter constants:")
for i in range(n):
    B[i] = float(input())

x = np.zeros(n)
tol = float(input("Enter tolerance: "))

while True:
    x_old = x.copy()

    for i in range(n):
        s = sum(A[i][j]*x[j] for j in range(n) if j != i)
        x[i] = (B[i] - s)/A[i][i]

    if np.allclose(x, x_old, atol=tol):
        break

print("Solution:", x)
```

# Code for Newton Interpolation
Used when interpolation point is near the beginning of the data.
Requires equally spaced data points.

```
import numpy as np

n = int(input("Enter number of data points: "))

x = list(map(float, input("Enter x values: ").split()))
y = list(map(float, input("Enter y values: ").split()))

diff = np.zeros((n,n))
diff[:,0] = y

for j in range(1,n):
    for i in range(n-j):
        diff[i][j] = diff[i+1][j-1] - diff[i][j-1]

value = float(input("Enter interpolation value: "))

h = x[1] - x[0]
p = (value - x[0]) / h

result = y[0]
fact = 1
p_term = 1

for i in range(1,n):
    p_term *= (p - i + 1)
    fact *= i
    result += (p_term * diff[0][i]) / fact

print("Interpolated value:", result)
```

# Code For Newton Backward Interpolation
Used when interpolation point is near the end of the data.

```
import numpy as np

n = int(input("Enter number of data points: "))

x = list(map(float, input("Enter x values: ").split()))
y = list(map(float, input("Enter y values: ").split()))

diff = np.zeros((n,n))
diff[:,0] = y

for j in range(1,n):
    for i in range(j,n):
        diff[i][j] = diff[i][j-1] - diff[i-1][j-1]

value = float(input("Enter interpolation value: "))

h = x[1] - x[0]
p = (value - x[-1]) / h

result = y[-1]
fact = 1
p_term = 1

for i in range(1,n):
    p_term *= (p + i - 1)
    fact *= i
    result += (p_term * diff[n-1][i]) / fact

print("Interpolated value:", result)
```
